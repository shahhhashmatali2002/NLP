```python
import pandas as pd
df = pd.read_csv('Twitter_Data.csv')
df.head()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
df['scores'] = df['clean_text'].apply(lambda clean_text: sid.polarity_scores(str(clean_text)))
df.head()
```

```python
df['compound'] = df['scores'].apply(lambda score_dict: score_dict['compound'])
df.head()
```

```python
df['label1'] = df['category'].apply(lambda c: 'pos' if c >= 0 else 'neg')
df.head()
```

```python
df['label2'] = df['compound'].apply(lambda c: 'pos' if c >= 0 else 'neg')
df.head()
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(classification_report(df['label1'],df['label2']))
```

```python
print(confusion_matrix(df['label1'],df['label2']))
```

```python
accuracy_score(df['label1'],df['label2'])
```
