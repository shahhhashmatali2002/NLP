```python
import pandas as pd
df = pd.read_csv('amazon_reviews.csv')
df.head()
```

```python
df['reviewText'].isnull().sum()
```

```python
df['reviewText'].isnull().sum()
```

```python
sid.polarity_scores(df.loc[1]['reviewText'])
```

```python
reviewerName = df['reviewText']
```

```python
for i in reviewerName:
#     sid.polarity_scores(i)
    if type(i) == str:
        gettingSentiments = sid.polarity_scores(i)
        print(gettingSentiments)
    
```

```python
df['scores'] = df['reviewText'].apply(lambda reviewText: sid.polarity_scores(str(reviewText)))
df.head()
```

```python
df['compound'] = df['scores'].apply(lambda score_dict: score_dict['compound'])
df.head()
```

```python
df['comp_score'] = df['compound'].apply(lambda c: 'pos' if c >= 0 else 'neg')
df.head()
```

```python
df['label'] = df['compound'].apply(lambda c: 'pos' if c >= 0 else 'neg')
df.head()
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(classification_report(df['label'],df['comp_score']))
```

```python
print(confusion_matrix(df['label'],df['comp_score']))
```

```python
accuracy_score(df['label'],df['comp_score'])
```
